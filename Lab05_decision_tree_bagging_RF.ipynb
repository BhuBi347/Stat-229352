```python
import pandas as pd
import graphviz

from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier

# import data
data = pd.read_csv("heart_disease.csv", na_values="?")
data.head()
```

```python
# split into X and y
y = data["label"]
X = data.drop("label", axis=1)

# split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# impute missing values
imputer = SimpleImputer(strategy="mean")
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Create a decision tree
clf = DecisionTreeClassifier()
```

![5CV](https://scikit-learn.org/stable/_images/grid_search_cross_validation.png)

```python
params = {'max_depth': [3, 6, 9, 12]}

gridcv = GridSearchCV(clf, params, scoring='accuracy', cv=5)
gridcv.fit(X_train, y_train)
```

```python
gridcv.best_estimator_
```

```python
plot_data = export_graphviz(gridcv.best_estimator_,
                            out_file=None,
                            filled=True,
                            rounded=True,
                            feature_names=data.columns[:-1],
                            class_names=['0', '1'])

graph = graphviz.Source(plot_data)
graph
```

## Bagged decision trees
* Bagging classifier ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html))

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import BaggingClassifier

bagged_clf = BaggingClassifier(
    estimator=clf, n_estimators=100, max_samples=0.8)
bagged_clf.fit(X_train, y_train)
print(classification_report(y_test, bagged_clf.predict(X_test)))
```

## Random forest classifier
* Random forest ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html))

```python
forest = RandomForestClassifier(n_estimators=200, max_depth=6)
forest.fit(X_train, y_train)
print(classification_report(y_test, forest.predict(X_test)))
```

**Decision tree Model**

```python
# Define the pipeline for Decision Tree
pipeline_dt = Pipeline([
    ('imputer', SimpleImputer(strategy="mean")),
    ('clf', DecisionTreeClassifier())
])

# Define the parameter grid for Decision Tree
param_grid_dt = {
    'clf__max_depth': [3, 5, 7, 10],
    'clf__min_samples_split': [2, 5, 10],
    'clf__min_samples_leaf': [1, 3, 5]
}
```

```python
# Instantiate GridSearchCV
gridcv_dt = GridSearchCV(pipeline_dt, param_grid_dt, scoring='accuracy', cv=5)

# Fit GridSearchCV to the training data
gridcv_dt.fit(X_train, y_train)
```

**Bagged decision trees grid search model**

```python
# Define the pipeline for Bagged Decision Trees
pipeline_bagging = Pipeline([
    ('imputer', SimpleImputer(strategy="mean")),
    ('bagged_trees', BaggingClassifier(estimator=DecisionTreeClassifier()))
])

# Define the parameter grid for Bagged Decision Trees
param_grid_bagging = {
    'bagged_trees__n_estimators': [50, 100, 150],
    'bagged_trees__max_samples': [0.6, 0.8, 1.0],
    'bagged_trees__max_features': [0.6, 0.8, 1.0]
}
```

```python
# Instantiate GridSearchCV
gridcv_bagging = GridSearchCV(pipeline_bagging, param_grid_bagging, scoring='accuracy', cv=5)

# Fit GridSearchCV to the training data
gridcv_bagging.fit(X_train, y_train)
```

**Random forest grid search Model**

```python
# Define the pipeline for Random Forest
pipeline_rf = Pipeline([
    ('imputer', SimpleImputer(strategy="mean")),
    ('rf', RandomForestClassifier())
])

# Define the parameter grid for Random Forest
param_grid_rf = {
    'rf__n_estimators': [50, 100, 150],
    'rf__max_depth': [3, 5, 7, 10],
    'rf__min_samples_leaf': [1, 3, 5]
}
```

```python
# Instantiate GridSearchCV
gridcv_rf = GridSearchCV(pipeline_rf, param_grid_rf, scoring='accuracy', cv=5)

# Fit GridSearchCV to the training data
gridcv_rf.fit(X_train, y_train)
```

**Model Compare**

```python
gridcv_dt_rp = classification_report(y_test, gridcv_dt.predict(X_test))
gridcv_bagging_rp = classification_report(y_test, gridcv_bagging.predict(X_test))
gridcv_rf_rp = classification_report(y_test, gridcv_rf.predict(X_test))

print("Decision Tree Report:")
print(gridcv_dt_rp)
print("------------------------------------------------------------")

print("\nBagged Decision Trees Report:")
print(gridcv_bagging_rp)
print("------------------------------------------------------------")

print("\nRandom Forest Report:")
print(gridcv_rf_rp)
```
